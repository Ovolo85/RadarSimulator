```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.arange(0, 4*np.pi, 0.02)
y = np.sin(x)
y_noise = []
for i in x:
    y_noise.append(np.sin(i) + np.random.normal(0, 0.1))
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(x, y_noise)
plt.tight_layout()
```

# Rolling Average

```python
n = 20
window = []
result = []

for data in y_noise:
    if len(window) == 0:
        window.append(data)
        result.append(data)
    else:
        if len(window)<n:
            window.append(data)
        else:
            window.append(data)
            window.pop(0)
        result.append(sum(window)/len(window))
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(x, y_noise, x, result)
plt.tight_layout()
```

# Low Pass

```python
alpha = 0.9
result = []

for data in y_noise:
    if len(result) == 0:
        result.append(data)
    else:
        result.append(alpha*result[-1] + (1-alpha)*data)
```

```python
fig, ax = plt.subplots(figsize=(10,7))
ax.plot(x, y_noise, x, result)
plt.tight_layout()
```
